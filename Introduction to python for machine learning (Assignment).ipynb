```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/HamoyeHQ/HDSC-Introduction-to-Python-for-machine-learning/files/7768140/FoodBalanceSheets_E_Africa_NOFLAG.csv'
df = pd.read_csv(url, encoding = 'latin-1')
```

```python
df.head()
```

```python
#Perform a groupby operation on ‘Element’.  What is the total number of the sum of Processing in 2017?
df.groupby('Element')['Y2017'].sum().reset_index()
```

```python
#What is the total number of unique countries in the dataset?
len(df['Area'].unique())
```

```python
#What is the total number of missing data in 2014?
df.isnull().sum()
```

```python
#What is the percentage of missing data in 2014 to 3 decimal places?
percent_missing = df.isnull().sum() * 100 / len(df)
print(round(percent_missing,3))
```

```python
#What is the total sum of Wine produced in 2015?
df.groupby('Item')['Y2015'].sum().reset_index()
```

```python
#What is the total sum of Wine produced in 2018 respectively?
df.groupby('Item')['Y2018'].sum().reset_index()
```

```python
#Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’. 
df.groupby('Area')['Y2017'].sum().reset_index()
```
